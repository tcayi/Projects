```python
#訓練和測試資料, 使用scikit-larn train_test_split()，預設比例 0.75:0.25
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(
iris_ds.data,iris_ds.target,random_state=0)
```

```python
#以視覺化(圖形)預先檢視資料
#以pd.DataFrame()將X_train(ndarray)轉成DataFrame
#以iris_ds.feature_names的字串作為欄位名稱

import pandas as pd
import matplotlib.pyplot as plt
import mglearn
%matplotlib inline

iris_df=pd.DataFrame(X_train,columns=iris_ds.feature_names)
iris_df.head()
```

```python
#從DataFrame 'iris_df' 建立pair圖,顏色依y_train的值
grr=pd.plotting.scatter_matrix(iris_df,c=y_train, figsize=(15,15),marker='x',
                     hist_kwds={'bins':20},s=60,alpha=0.7,cmap=mglearn.cm3)
```

```python
#建立kNN模型
from sklearn.neighbors import KNeighborsClassifier
knnModel=KNeighborsClassifier(n_neighbors=1) #k=1
```

```python
#以train datasets 做 fit
knnModel.fit(X_train,y_train)
```

```python
#預測, 以X_test[0], 並以y_test[0] 對答案
import numpy as np
newp=np.array([5.4,3.2,2.5,0.8])
newp=newp.reshape(1,-1)
```

```python
knnModel.predict([[5.4,3.2,2.5,0.8]])
```

```python
#預測,新採集的一朵鳶尾花樣本 尺寸為 (5,2.9,1,0.2)
#將list轉為shape:(1,4) 的 ndarray
sample_01=[5,2.9,1,0.2] 
new_array_01=np.array(sample_01).reshape(1,-1)
prediction_01=knnModel.predict(new_array_01)
```

```python
#以zip編一個鳶尾花品種名稱字典 "dict_iris"
dict_iris={no:'Iris-'+name for no,name in zip(range(0,3),iris_ds.target_names)}
```

```python
#印出預測鳶尾花樣本的品種
print("鳶尾花尺寸:")
print("花萼長:",sample_01[0],"cm |"," 花萼寬:",sample_01[1],"cm |",
     "花瓣長:",sample_01[2],"cm |"," 花瓣寬:",sample_01[2],"cm |","\n")
print("預測品種是:",dict_iris[prediction_01[0]])
```

```python
#評估模型準確率
#使用 model.score() 計算
knnModel.score(X_test,y_test)
```

```python
#整合成一個生產系統
#以joblib.dump() 將模型 knnModel dump 到現行目錄下,檔名 'iris_knn.pkl'
import joblib #import joblib module
joblib.dump(knnModel,'iris_knn.pkl')
```

```python
#將 model file 'iris_knn.pkl' 以 joblib.load() 載入
loadKnnModel=joblib.load('iris_knn.pkl')
```

```python
#以載入的 model "loadKnnModel" 對 new_array_01 做預測
dict_iris[loadKnnModel.predict(new_array_01)[0]]
```
